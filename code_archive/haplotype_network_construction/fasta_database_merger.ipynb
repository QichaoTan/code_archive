```python
from Bio import SeqIO
```

```python
PATH_database_fasta = '/home/qichaotan/CODE_ARCHIVE/code_archive/haplotype_network_construction/sequence.fasta'
PATH_data_fasta = '/home/qichaotan/CODE_ARCHIVE/code_archive/haplotype_network_construction/COX3_ASV_hard.fasta'
PATH_output = '/home/qichaotan/CODE_ARCHIVE/code_archive/haplotype_network_construction/Curated_fasta.fasta'
forward_primer = 'TTGCTTGGGTTTTTCAGCGT'
reverse_primer = 'GCCGCAAGTTTAACTCCAACA'
```

```python
count_overall, count_database, count_data = 0, 0, 0
record_set = set()
with open(PATH_output,'w') as output_handle:
    records = SeqIO.parse(PATH_database_fasta,'fasta')
    for record in records:
        record.id = record.description.split(' ')[4]
        if record.seq.find(forward_primer) == -1 or record.seq.find(reverse_primer) == -1:
            continue
        record.seq = record.seq[record.seq.find(forward_primer):int(record.seq.find(reverse_primer)) + len(reverse_primer)]
        

        if record.seq not in record_set:
            SeqIO.write(record,output_handle,'fasta')
            count_overall += 1
            count_database += 1
        record_set.add(record.seq)

with open(PATH_output,'a') as output_handle:
    records = SeqIO.parse(PATH_data_fasta,'fasta')
    for record in records:
        count_overall += 1
        count_data += 1
        SeqIO.write(record,output_handle,'fasta')

print(f'{count_database} dereped records are in database, {count_data} records are in data, adding to {count_overall} records in the curated fasta')
```
